```python
import requests

competition_slug = 'datathon-2026-round-1'

try:
    download_path = kagglehub.competition_download(competition_slug)
    print('Path to competition files:', download_path)
except requests.HTTPError as e:
    status = e.response.status_code if e.response is not None else None
    msg = [f'Kaggle download failed with HTTP status: {status}']
    if status in (401, 403):
        msg.append('Possible fixes:')
        msg.append('1) Regenerate Kaggle API token/key and update .env')
        msg.append('2) Accept competition rules on the competition page while logged in')
        msg.append('3) Restart notebook kernel so new env vars are loaded')
    raise RuntimeError('\n'.join(msg)) from e
```

```python
p = Path(download_path)
print('Exists:', p.exists())
print('Files:')
for file in sorted(p.rglob('*')):
    if file.is_file():
        print('-', file.relative_to(p))
```
